### 加载模型和分词器

```python
from transformers import AutoTokenizer
import transformers
import torch


model_id = "codellama/CodeLlama-7b-hf"
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

### 准备通道

```python
pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    torch_dtype=torch.float16,
    device_map="auto",
)
```

### 生成代码

```python
def generate_code(prompt):
    sequences = pipeline(
        prompt,
        do_sample=True,
        temperature=0.1,
        top_p=0.9,
        num_return_sequences=1,
        eos_token_id=tokenizer.eos_token_id,
        max_length=128,
    )
    for seq in sequences:
        print(f"Result: {seq['generated_text']}")
```

### 测试多个查询

```python
generate_code("def fibonacci(")
```

```python
generate_code("def factorial(")
```

```python
generate_code("def remove_last_word(")
```

```python
generate_code("def remove_non_ascii(s: str) -> str:")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import transformers
import torch

model_id = "codellama/CodeLlama-7b-hf"
tokenizer2 = AutoTokenizer.from_pretrained(model_id)
model2 = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16
).to("cuda")


```

```python
prompt = '''def remove_non_ascii(s: str) -> str:
    """ <FILL_ME>
    return result
'''

input_ids = tokenizer2(prompt, return_tensors="pt")["input_ids"].to("cuda")
output = model2.generate(
    input_ids,
    max_new_tokens=200,
)
output = output[0].to("cpu")

filling = tokenizer.decode(output[input_ids.shape[1]:], skip_special_tokens=True)
```

```python
print(prompt.replace("<FILL_ME>", filling))
```
